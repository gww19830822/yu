### 拓展延申：列表拷贝问题(*)

比较以下两种复制列表的方式的区别。

````python
#直接赋值
l1 = [1,2,3]
l2 = l1
l2[1] = 4
print(l1)
print(l2)
````

````python
#切片赋值
l1 = [1,2,3]
l2 = l1[:]
l2[1] = 4
print(l1)
print(l2)
````
